# ブロッホ球

量子ビットの状態を表現するための有用な幾何学的手法として、*ブロッホ球（Bloch sphere）* と呼ばれるものがあります。
これは非常に便利な表現ですが、残念ながら量子ビットに対してのみ有効です。系が3つ以上の古典状態をもつ場合には、対応する表現はもはや球面にはなりません。

## 球面上の点としての量子ビット状態

まず、量子ビットの量子状態ベクトル $\alpha \vert 0\rangle + \beta \vert 1\rangle$ について考えましょう。
すべての量子ビットの状態ベクトルはグローバル位相を除いて同値であり、$\alpha \geq 0$を満たす形にできるため、 $\alpha$ を非負の実数とするベクトルに限定して考えることができます。
このことにより、次のように書くことができます：

$$
\vert\psi\rangle = \cos\bigl(\theta/2\bigr) \vert 0\rangle + e^{i\phi} \sin\bigl(\theta/2\bigr) \vert 1\rangle
$$

ここで、$\theta \in [0,\pi]$ 、 $\phi\in[0,2\pi)$は実数です。
$\theta$を$0$ から$\pi$ の範囲に取り、さらに正弦と余弦の引数を $1/2$ しているのは、この種のベクトルを表現する際の慣習であり、後の議論を少し簡単にするためです。

さて、与えられた量子状態ベクトル$\alpha \vert 0\rangle + \beta \vert 1\rangle$ に対して、$\theta$と$\phi$ が完全に一意に定まるわけではありませんが、ほぼ一意に定まります。
特に、$\beta = 0$ の場合には $\theta = 0$ となり、このとき $\phi$の値は何であっても結果に影響しないため、任意に選ぶことができます。
同様に、$\alpha = 0$ の場合には$\theta = \pi$ となり、この場合も $\phi$ は本質的に無関係です (任意の$\phi$に対して、状態はグローバル位相を除いて $e^{i\phi}\vert 1\rangle$ と同値になります)。
一方、 $\alpha$ と $\beta$ のどちらも 0 でない場合には、グローバル位相を除いて $\vert\psi\rangle$が$\alpha\vert 0\rangle + \beta\vert 1\rangle$と一致するような $(\theta,\phi)$ の組は一意に定まります。

次に、この状態の密度行列表現について考えましょう。

$$
\vert\psi\rangle\langle\psi\vert
= \begin{pmatrix}
\cos^2(\theta/2) & e^{-i\phi}\cos(\theta/2)\sin(\theta/2)\\[2mm]
e^{i\phi}\cos(\theta/2)\sin(\theta/2) & \sin^2(\theta/2)
\end{pmatrix}
$$

いくつかの三角関数の恒等式

$$
\begin{gathered}
\cos^2(\theta/2) = \frac{1 + \cos(\theta)}{2},\\[2mm]
\sin^2(\theta/2) = \frac{1 - \cos(\theta)}{2},\\[2mm]
\cos(\theta/2) \sin(\theta/2) = \frac{\sin(\theta)}{2},
\end{gathered}
$$

および公式 $e^{i\phi} = \cos(\phi) + i\sin(\phi)$ を用いると、密度行列を次のように簡単化できます。

$$
\vert\psi\rangle\langle\psi\vert
= \frac{1}{2}
\begin{pmatrix}
1 + \cos(\theta) & (\cos(\phi) - i \sin(\phi)) \sin(\theta)\\[1mm]
(\cos(\phi) + i \sin(\phi)) \sin(\theta) & 1 - \cos(\theta)
\end{pmatrix}
$$

これにより、この密度行列をパウリ行列の線形結合として簡単に表すことができます：

$$
  \mathbb{I} =
  \begin{pmatrix}
    1 & 0\\[1mm]
    0 & 1
  \end{pmatrix},
  \quad
  \sigma_x =
  \begin{pmatrix}
    0 & 1\\[1mm]
    1 & 0
  \end{pmatrix},
  \quad
  \sigma_y =
  \begin{pmatrix}
    0 & -i\\[1mm]
    i & 0
  \end{pmatrix},
  \quad
  \sigma_z =
  \begin{pmatrix}
    1 & 0\\[1mm]
    0 & -1
  \end{pmatrix}.
$$

具体的には、次の式が得られます：

$$
\vert\psi\rangle\langle\psi\vert
= \frac{\mathbb{I} + \sin(\theta) \cos(\phi)\sigma_x + \sin(\theta)\sin(\phi) \sigma_y + \cos(\theta) \sigma_z}{2}.
$$

この式の分子における $\sigma_x,$ $\sigma_y,$  $\sigma_z$ の係数はすべて実数なので、それらをまとめて通常の3次元ユークリッド空間内のベクトルとして表すことができます。

$$
\bigl(\sin(\theta) \cos(\phi), \sin(\theta)\sin(\phi), \cos(\theta)\bigr)
$$

実際、これは単位ベクトルになっています。
これは*球面座標*を用いると $(1,\theta,\phi)$と書くことができます。
最初の座標である $1$ は*半径*（あるいは*動径距離*）を表し（この場合は常に $1$）、$\theta$  は*極角*、$\phi$ は*方位角*を表します。

言葉で説明すると、球を地球に見立てたとき、極角 $\theta$ は北極から南へどれだけ回転してその点に到達するかを表し、その範囲は $0$ から $\pi = 180^{\circ}$ です。一方、方位角  $\phi$ は本初子午線から東へどれだけ回転するかを表し、その範囲は $0$ から $2\pi = 360^{\circ}$です。
ここでは、本初子午線を正の $x$軸を通り、球面上で一方の極からもう一方の極へとつながる曲線と定義しています。

球面上の任意の点はこのように表すことができます。ー すなわち、量子ビットのすべての純粋状態に対応する点は、ちょうど$3$次元実空間内の球面に一致します。（この球面は通常、表面が2次元であることから*単位$2$球面*と呼ばれます。）

この単位$2$球面上の点と量子ビットの純粋状態を対応付けることで、これらの状態の*ブロッホ球*表現が得られます。

## 6つの重要な例

1. *標準基底* $\{\vert 0\rangle,\vert 1\rangle\}.$
   まず状態 $\vert 0\rangle$から始めましょう。
   その密度行列は次のように書けます：

   $$
   \vert 0 \rangle \langle 0 \vert = \frac{\mathbb{I} + \sigma_z}{2}
   $$

   分子に現れるパウリ行列の係数をまとめると、対応する単位$2$球面上の点（デカルト座標）は$(0,0,1)$になります。
   球面座標では $(1,0,\phi)$と表され、$\phi$ は任意の角度を取れます。

   これは次の表式とも一致しています：

   $$
   \vert 0\rangle = \cos(0) \vert 0\rangle + e^{i \phi} \sin(0) \vert 1\rangle,
   $$

   この式も任意の $\phi$に対して成り立ちます。
   直感的には、極角 $\theta=0$ なので、ブロッホ球の北極に位置しており、このとき方位角$\phi$ は意味を持ちません。

   同様に、状態$\vert 1\rangle$ の密度行列は

   $$
   \vert 1 \rangle \langle 1 \vert = \frac{\mathbb{I} - \sigma_z}{2}
   $$

   と書けます。この場合のデカルト座標は $(0,0,-1)$、球面座標では $(1,\pi,\phi)$ ( $\phi$ は任意)です。 極角が$\pi$ なので、ブロッホ球の南極に対応し、ここでも方位角は重要ではありません。

3. *基底 $\{\vert + \rangle, \vert - \rangle\}.$*
   これらの状態に対応する密度行列は次の通りです：

   $$
   \begin{aligned}
   \vert {+} \rangle\langle {+} \vert & = \frac{\mathbb{I} + \sigma_x}{2}\\[2mm]
   \vert {-} \rangle\langle {-} \vert & = \frac{\mathbb{I} - \sigma_x}{2}
   \end{aligned}
   $$

   対応する単位$2$球面上の点は、デカルト座標で$(1,0,0)$と $(-1,0,0)$、
   球面座標ではそれぞれ $(1,\pi/2,0)$ と $(1,\pi/2,\pi)$ です。

   言い換えると、$\vert +\rangle$ cは正の $x$軸が球面と交わる点、$\vert -\rangle$ は負の$x$軸との交点に対応します。 より直感的には、$\vert +\rangle$ はブロッホ球の赤道上で本初子午線と交わる点、 $\vert - \rangle$ はその反対側の赤道上の点です。

   
4. *基底* $\{\vert {+i} \rangle, \vert {-i} \rangle\}.$
   前のレッスンで学んだように、これらの状態は次のように定義されます：

   $$
   \begin{aligned}
   \vert {+i} \rangle & = \frac{1}{\sqrt{2}} \vert 0 \rangle + \frac{i}{\sqrt{2}} \vert 1 \rangle\\[2mm]
   \vert {-i} \rangle & = \frac{1}{\sqrt{2}} \vert 0 \rangle - \frac{i}{\sqrt{2}} \vert 1 \rangle.
   \end{aligned}
   $$

   このとき密度行列は

   $$
   \begin{aligned}
   \vert {+i} \rangle\langle {+i} \vert & = \frac{\mathbb{I} + \sigma_y}{2}\\[2mm]
   \vert {-i} \rangle\langle {-i} \vert & = \frac{\mathbb{I} - \sigma_y}{2}
   \end{aligned}
   $$

   となります。
   単位$2$球面上の対応する点は、デカルト座標で $(0,1,0)$ と$(0,-1,0)$、球面座標ではそれぞれ $(1,\pi/2,\pi/2)$と $(1,\pi/2,3\pi/2)$です。

   言い換えると、 $\vert {+i} \rangle$ は正の $y$軸との交点、$\vert {-i} \rangle$ tは負の $y$軸との交点に対応します。

![Illustration of six examples of pure states on the Bloch sphere](https://quantum.cloud.ibm.com/assets-docs-learning/_next/image?url=%2Flearning%2Fimages%2Fcourses%2Fgeneral-formulation-of-quantum-information%2Fdensity-matrices%2Fsix-Bloch-sphere-points.avif&w=3840&q=75)

このシリーズの中で何度も登場している別のクラスの量子状態ベクトルとして、次のものがあります：

$$
\vert \psi_{\alpha} \rangle = \cos(\alpha) \vert 0\rangle + \sin(\alpha) \vert 1\rangle \qquad \text{(for $\alpha \in [0,\pi)$)}
$$

これらの状態の密度行列は次のように表されます：

$$
\vert \psi_{\alpha} \rangle \langle \psi_{\alpha} \vert =
\begin{pmatrix}
\cos^2(\alpha) & \cos(\alpha)\sin(\alpha)\\[2mm]
\cos(\alpha)\sin(\alpha) & \sin^2(\alpha)
\end{pmatrix}
= \frac{\mathbb{I} + \sin(2\alpha) \sigma_x + \cos(2\alpha) \sigma_z}{2}
$$

## 点の凸結合

これまで密度行列について見てきたのと同様に、ブロッホ球上の点の凸結合をとることで、量子ビットの密度行列を表現することができます。
一般には、その結果はブロッホ球の*内部*の点となり、これは純粋状態ではない状態（混合状態）の密度行列を表します。
ブロッホ球の内部の点も含めて明示的に扱う場合には、*ブロッホ球体（Bloch ball）*　という言い方をすることもあります。

例えば、量子ビットの完全混合状態を表す密度行列 $\frac{1}{2}\mathbb{I}$ は、次の2通りの方法で表せることを見てきました：
$$
\frac{1}{2} \mathbb{I} = \frac{1}{2} \vert 0\rangle\langle 0\vert + \frac{1}{2} \vert 1\rangle\langle 1\vert
\quad\text{and}\quad
\frac{1}{2} \mathbb{I} = \frac{1}{2} \vert +\rangle\langle +\vert + \frac{1}{2} \vert -\rangle\langle -\vert.
$$

さらに、

$$
\frac{1}{2} \mathbb{I} =
\frac{1}{2} \vert {+i}\rangle\langle {+i} \vert
+ \frac{1}{2} \vert {-i} \rangle\langle {-i}\vert,
$$


とも書けます。
より一般には、任意の2つの直交する量子ビット状態（これはブロッホ球上の互いに正反対の点に対応します）を用いることができます。
ブロッホ球上でこれらの点を同様に平均すると、常に同じ点、すなわち球の中心が得られます。
これは次の式とも一致しています：

$$
\frac{1}{2} \mathbb{I} = \frac{\mathbb{I} + 0 \cdot \sigma_x + 0 \cdot \sigma_y + 0 \cdot \sigma_z}{2},
$$

したがって、対応するデカルト座標は$(0,0,0)$となります。

ブロッホ球上の点の凸結合の別の例として、前の節で扱った次のものがあります：

$$
\frac{1}{2} \vert 0\rangle\langle 0 \vert + \frac{1}{2} \vert +\rangle\langle + \vert
= \begin{pmatrix} \frac{3}{4} & \frac{1}{4}\\[2mm]
\frac{1}{4} & \frac{1}{4}
\end{pmatrix}
= \cos^2(\pi/8) \vert \psi_{\pi/8} \rangle \langle \psi_{\pi/8}\vert +
\sin^2(\pi/8) \vert \psi_{5\pi/8} \rangle \langle \psi_{5\pi/8}\vert
$$
