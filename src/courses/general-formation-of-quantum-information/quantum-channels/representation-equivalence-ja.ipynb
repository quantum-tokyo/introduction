# 表現の同値性

これまでに、チャネルを数学的に表現する3つの異なる方法、すなわちスタインスプリング表現、クラウス表現、チョイ表現について説明してきました。
また、チャネルの定義もあります。それは、チャネルとは、複合系の一部にだけ適用された場合でも常に密度行列を密度行列に変換する線形写像である、というものです。
この講義の残りの部分は、これら3つの表現が同値であり、定義を正確に捉えていることの数学的な証明に充てられます。

## 証明の概要

私たちの目標は、4つの命題からなる集まりの同値性を確立することです。まずはそれらを正確に書き下すことから始めましょう。
4つの命題はいずれも、この講義を通して用いてきた慣例に従います。すなわち、$\Phi$ は正方行列から正方行列への線形写像であり、入力行列の行と列は系 $\mathsf{X}$（入力系）の古典的状態に対応させられ、出力行列の行と列は系 $\mathsf{Y}$（出力系）の古典的状態に対応させられています。

1. $\Phi$ は $\mathsf{X}$ から $\mathsf{Y}$ へのチャネルである。すなわち、$\Phi$ は、より大きな複合系の一部に作用する場合でも、常に密度行列を密度行列に変換する。

2. チョイ行列 $J(\Phi)$ は半正定値であり、条件 $\operatorname{Tr}_{\mathsf{Y}}(J(\Phi)) = \mathbb{I}_{\mathsf{X}}$ を満たす。

3. $\Phi$ にはクラウス表現が存在する。すなわち、あらゆる入力 $\rho$ に対して式 $\Phi(\rho) = \sum_{k = 0}^{N-1} A_k \rho A_k^{\dagger}$ が真となり、かつ条件 $\sum_{k = 0}^{N-1} A_k^{\dagger} A_k = \mathbb{I}_{\mathsf{X}}$ を満たす行列 $A_0,\ldots,A_{N-1}$ が存在する。

4. $\Phi$ にはスタインスプリング表現が存在する。すなわち、組 $(\mathsf{W},\mathsf{X})$ と $(\mathsf{G},\mathsf{Y})$ が同じ数の古典的状態を持つような系 $\mathsf{W}$ と $\mathsf{G}$、および $(\mathsf{W},\mathsf{X})$ から $(\mathsf{G},\mathsf{Y})$ へのユニタリ操作を表すユニタリ行列 $U$ が存在して、$\Phi(\rho) = \operatorname{Tr}_{\mathsf{G}}\bigl( U (\vert 0\rangle\langle 0 \vert \otimes \rho) U^{\dagger} \bigr)$ が成り立つ。

証明の進め方は、含意の巡回（サイクル）を証明するというものです。
すなわち、リストの1番目の命題が2番目を含意し、2番目が3番目を含意し、3番目が4番目を含意し、4番目の命題が1番目を含意することを示します。
これにより4つの命題がすべて同値であること——つまり、与えられた $\Phi$ に対してそれらがすべて真であるか、すべて偽であるかのいずれかであること——が確立されます。なぜなら、含意はどの命題からどの命題へも推移的にたどれるからです。

これは、複数の命題が同値であることを証明するときの一般的な戦略であり、そのような文脈で使える便利なコツは、できるだけ証明しやすくなるように含意を設定することです。
ここではまさにそうしています——そして実際、4つの含意のうち2つはすでに登場しています。

## チャネルからチョイ行列へ

上に挙げた命題を番号で参照すると、最初に証明する含意は 1 $\Rightarrow$ 2 です。
この含意は、チャネルのチョイ状態の文脈ですでに議論しました。
ここでは数学的な詳細をまとめます。

入力系 $\mathsf{X}$ の古典的状態集合を $\Sigma$ とし、$n = \vert\Sigma\vert$ とします。
$\mathsf{X}$ の2つのコピーが一緒に状態

$$
\vert \psi \rangle = \frac{1}{\sqrt{n}} \sum_{a \in \Sigma} \vert a \rangle \otimes \vert a \rangle,
$$

にあり、その2番目のコピーに $\Phi$ を適用する状況を考えます。密度行列として、これは次で与えられます。

$$
\vert \psi \rangle \langle \psi \vert = \frac{1}{n} \sum_{a,b \in \Sigma}
\vert a\rangle\langle b \vert \otimes \vert a\rangle\langle b \vert.
$$

結果は次のように書けます。

$$
(\operatorname{Id}\otimes \,\Phi) \bigl(\vert \psi \rangle \langle \psi \vert\bigr)
= \frac{1}{n} \sum_{a,b = 0}^{n-1} \vert a\rangle\langle b \vert \otimes \Phi\bigl(\vert a\rangle\langle b \vert\bigr)
= \frac{J(\Phi)}{n},
$$

そして $\Phi$ がチャネルであるという仮定により、これは密度行列でなければなりません。
すべての密度行列と同様にこれは半正定値でなければならず、半正定値行列に正の実数を掛けると別の半正定値行列が得られるので、$J(\Phi) \geq 0$ となります。

さらに、$\Phi$ がチャネルであるという仮定の下では、トレースを保存しなければならないので、

$$
\begin{aligned}
\operatorname{Tr}_{\mathsf{Y}} (J(\Phi))
& = \sum_{a,b\in\Sigma} \operatorname{Tr}\bigl(\Phi( \vert a\rangle\langle b \vert)\bigr) \, \vert a\rangle\langle b \vert\\
& = \sum_{a,b\in\Sigma} \operatorname{Tr}\bigl(\vert a\rangle\langle b \vert\bigr) \, \vert a\rangle\langle b \vert\\
& = \sum_{a\in\Sigma} \vert a\rangle\langle a \vert\\
& = \mathbb{I}_{\mathsf{X}}.
\end{aligned}
$$

## チョイ表現からクラウス表現へ

2番目の含意は、再びリストの命題を番号で参照すると、2 $\Rightarrow$ 3 です。
明確にしておくと、ここでは他の命題は無視します——特に、$\Phi$ がチャネルであるという仮定を置くことはできません。
使えるのは、$\Phi$ が線形写像であり、そのチョイ表現が $J(\Phi) \geq 0$ と $\operatorname{Tr}_{\mathsf{Y}} (J(\Phi)) = \mathbb{I}_{\mathsf{X}}$ を満たす、ということだけです。

しかし、これだけあれば、$\Phi$ が条件

$$
\sum_{k = 0}^{N-1} A_k^{\dagger} A_k = \mathbb{I}_{\mathsf{X}}
$$

を満たすクラウス表現

$$
\Phi(\rho) = \sum_{k = 0}^{N-1} A_k \rho A_k^{\dagger}
$$

を持つと結論するのに十分です。

まず、$J(\Phi)$ が半正定値であるという決定的に重要な仮定から始めます。これは、$J(\Phi)$ をベクトル $\vert\psi_0\rangle,\ldots,\vert\psi_{N-1}\rangle$ の何らかの選び方によって次の形で表せることを意味します。

$$
J(\Phi) = \sum_{k = 0}^{N-1} \vert \psi_k \rangle \langle \psi_k \vert
\tag{1}
$$

一般に、これを行う方法は複数あります——そして実際、これは $\Phi$ のクラウス表現を選ぶ際の自由度を直接反映しています。

そのような表式を得る一つの方法は、まずスペクトル定理を用いて次のように書くことです。

$$
J(\Phi) = \sum_{k = 0}^{N-1} \lambda_k \vert \gamma_k \rangle \langle \gamma_k \vert,
$$

ここで $\lambda_0,\ldots,\lambda_{N-1}$ は $J(\Phi)$ の固有値（$J(\Phi)$ が半正定値であるため、必ず非負の実数です）、$\vert\gamma_0\rangle,\ldots,\vert\gamma_{N-1}\rangle$ は固有値 $\lambda_0,\ldots,\lambda_{N-1}$ に対応する単位固有ベクトルです。

固有値の選び方には（順序を除いて）自由度がありませんが、固有ベクトルの選び方には自由度があることに注意してください。特に、重複度が1より大きい固有値がある場合はそうです。
したがって、これは $J(\Phi)$ の一意な表式ではありません——単にそのような表式が一つあると仮定しているだけです。
いずれにせよ、固有値は非負の実数なので、非負の平方根を持ちます。よって、各 $k = 0,\ldots,N-1$ に対して

$$
\vert\psi_k\rangle = \sqrt{\lambda_k} \vert \gamma_k\rangle
$$

を選ぶことで、形 $(1)$ の表式が得られます。

ただし、表式 $(1)$ がこのようにスペクトル分解から得られることは本質的ではなく、特にベクトル $\vert\psi_0\rangle,\ldots,\vert\psi_{N-1}\rangle$ は一般には直交している必要はありません。
とはいえ、望むならこれらのベクトルを直交するように選べること、さらに $N$ を $nm$ より大きくする必要は決してないこと（$n$ と $m$ はそれぞれ $\mathsf{X}$ と $\mathsf{Y}$ の古典的状態の数を表すことを思い出してください）は注目に値します。

次に、ベクトル $\vert\psi_0\rangle,\ldots,\vert\psi_{N-1}\rangle$ のそれぞれをさらに次のように分解できます。

$$
\vert\psi_k\rangle = \sum_{a\in\Sigma} \vert a\rangle \otimes \vert \phi_{k,a}\rangle,
$$

ここでベクトル $\{ \vert \phi_{k,a}\rangle \}$ は $\mathsf{Y}$ の古典的状態に対応する成分を持ち、各 $a\in\Sigma$ と $k=0,\ldots,N-1$ に対して次の式で明示的に決まります。

$$
\vert \phi_{k,a}\rangle = \bigl( \langle a \vert \otimes \mathbb{I}_{\mathsf{Y}}\bigr) \vert \psi_k\rangle
$$

$\vert\psi_0\rangle,\ldots,\vert\psi_{N-1}\rangle$ は必ずしも単位ベクトルではありませんが、これは、組 $(\mathsf{X},\mathsf{Y})$ の量子状態ベクトルが与えられたときに系 $\mathsf{X}$ に対して標準基底測定を行うと何が起こるかを分析するのに用いるのと同じ手続きです。

そしてここで、この部分の証明を成立させるコツにたどり着きます。
クラウス行列 $A_0,\ldots,A_{N-1}$ を次の式に従って定義します。

$$
A_k = \sum_{a\in\Sigma} \vert \phi_{k,a}\rangle\langle a \vert
$$

この式を純粋に記号的に考えることができます。$\vert a\rangle$ は事実上ひっくり返されて $\langle a\vert$ になり、右側に移されて行列を形成します。
証明を検証する目的では、この式だけあれば十分です。

しかし、ベクトル $\vert\psi_k\rangle$ と行列 $A_k$ の間には単純で直観的な関係があります。それは、$A_k$ を *ベクトル化* すると $\vert\psi_k\rangle$ が得られる、というものです。
$A_k$ をベクトル化するとは、列を上から下へ積み重ねて（一番左の列を一番上に、一番右の列を一番下にして）ベクトルを形成することを意味します。
たとえば、$\mathsf{X}$ と $\mathsf{Y}$ がともに量子ビットで、ある $k$ の選び方に対して

$$
\begin{aligned}
\vert\psi_k\rangle & = \alpha_{00} \vert 0\rangle \otimes \vert 0\rangle +
\alpha_{01} \vert 0\rangle \otimes \vert 1\rangle +
\alpha_{10} \vert 1\rangle \otimes \vert 0\rangle +
\alpha_{11} \vert 1\rangle \otimes \vert 1\rangle\\[2mm]
& = \begin{pmatrix}
\alpha_{00} \\[1mm]
\alpha_{01} \\[1mm]
\alpha_{10} \\[1mm]
\alpha_{11}
\end{pmatrix},
\end{aligned}
$$

であれば、次のようになります。

$$
\begin{aligned}
A_k & = \alpha_{00} \vert 0\rangle\langle 0\vert +
\alpha_{01} \vert 1\rangle\langle 0\vert +
\alpha_{10} \vert 0\rangle\langle 1\vert +
\alpha_{11} \vert 1\rangle\langle 1\vert\\[2mm]
& = \begin{pmatrix}
\alpha_{00} & \alpha_{10}\\[1mm]
\alpha_{01} & \alpha_{11}
\end{pmatrix}.
\end{aligned}
$$

（注意：行列のベクトル化は、行列の *行* を転置して上から下へ積み重ねて列ベクトルを形成する、というやや異なる方法で定義されることもあります。）

まず、このクラウス行列の選び方が写像 $\Phi$ を正しく記述することを検証し、その後で残りの必要条件を検証します。
話を整理するために、新しい写像 $\Psi$ を次のように定義しましょう。

$$
\Psi(\rho) = \sum_{k = 0}^{N-1} A_k \rho A_k^{\dagger}
$$

したがって、私たちの目標は $\Psi = \Phi$ を検証することです。

これを行う方法は、これらの写像のチョイ表現を比較することです。
チョイ表現は忠実なので、$J(\Phi) = J(\Psi)$ であるとき、かつそのときに限り $\Psi = \Phi$ が成り立ちます。
ここで、表式

$$
\vert\psi_k\rangle = \sum_{a\in\Sigma} \vert a\rangle \otimes \vert \phi_{k,a}\rangle
\quad\text{および}\quad
A_k = \sum_{a\in\Sigma} \vert \phi_{k,a}\rangle\langle a \vert
$$

とテンソル積の双線形性を用いて $J(\Psi)$ を計算し、単純化するだけです。

$$
\begin{aligned}
J(\Psi) & = \sum_{a,b\in\Sigma}  \vert a\rangle \langle b \vert \otimes \sum_{k = 0}^{N-1} A_k \vert a\rangle \langle b \vert A_k^{\dagger}\\[2mm]
& = \sum_{a,b\in\Sigma}  \vert a\rangle \langle b \vert \otimes \sum_{k = 0}^{N-1}  \vert \phi_{k,a} \rangle \langle \phi_{k,b} \vert \\[2mm]
& = \sum_{k = 0}^{N-1} \biggl(\sum_{a\in\Sigma} \vert a\rangle \otimes \vert \phi_{k,a} \rangle\biggr)
\biggl(\sum_{b\in\Sigma} \langle b\vert \otimes \langle \phi_{k,b} \vert\biggr)\\[2mm]
& = \sum_{k = 0}^{N-1} \vert \psi_k \rangle \langle \psi_k \vert \\[2mm]
& = J(\Phi)
\end{aligned}
$$

したがって、私たちのクラウス行列は $\Phi$ を正しく記述しています。

残るは $A_0,\ldots,A_{N-1}$ に対する必要条件を確認することです。これは（まだ使っていない）仮定 $\operatorname{Tr}_{\mathsf{Y}}(J(\Phi)) = \mathbb{I}_{\mathsf{X}}$ と等価であることがわかります。
示すのは次の関係です。

$$
\Biggl( \sum_{k = 0}^{N-1} A_k^{\dagger} A_k \Biggr)^{T} = \operatorname{Tr}_{\mathsf{Y}}(J(\Phi))
\tag{2}
$$

（ここで左辺は *行列の転置* を指しています。）

左辺から始めると、まず次のことがわかります。

$$
\begin{aligned}
\Biggl(\sum_{k = 0}^{N-1} A_k^{\dagger} A_k\Biggr)^T
& = \Biggl(\sum_{k = 0}^{N-1} \sum_{a,b\in\Sigma} \vert b \rangle \langle \phi_{k,b} \vert \phi_{k,a} \rangle \langle a \vert\Biggr)^T\\
& = \sum_{k = 0}^{N-1} \sum_{a,b\in\Sigma} \langle \phi_{k,b} \vert \phi_{k,a} \rangle \vert a \rangle  \langle b \vert.
\end{aligned}
$$

最後の等式は、転置が線形であり $\vert b\rangle\langle a \vert$ を $\vert a\rangle\langle b \vert$ に写すという事実から従います。

式の右辺に移ると、

$$
J(\Phi) = \sum_{k = 0}^{N-1} \vert \psi_k\rangle\langle\psi_k \vert
= \sum_{k = 0}^{N-1} \sum_{a,b\in\Sigma} \vert a\rangle \langle b \vert \otimes \vert\phi_{k,a}\rangle\langle \phi_{k,b} \vert
$$

であり、したがって

$$
\begin{aligned}
\operatorname{Tr}_{\mathsf{Y}}(J(\Phi))
& = \sum_{k = 0}^{N-1} \sum_{a,b\in\Sigma} \operatorname{Tr}\bigl(\vert\phi_{k,a}\rangle\langle \phi_{k,b} \vert \bigr)\,
\vert a\rangle \langle b \vert\\
& = \sum_{k = 0}^{N-1} \sum_{a,b\in\Sigma} \langle \phi_{k,b} \vert \phi_{k,a} \rangle \vert a \rangle  \langle b \vert.
\end{aligned}
$$

同じ結果が得られたので、式 $(2)$ が検証されました。
仮定 $\operatorname{Tr}_{\mathsf{Y}} (J(\Phi)) = \mathbb{I}_{\mathsf{X}}$ により、

$$
\Biggl(\sum_{k = 0}^{N-1} A_k^{\dagger} A_k\Biggr)^T = \mathbb{I}_{\mathsf{X}}
$$

が従い、したがって、単位行列はそれ自身の転置であるため、必要条件が成り立ちます。

$$
\sum_{k = 0}^{N-1} A_k^{\dagger} A_k = \mathbb{I}_{\mathsf{X}}
$$

## クラウス表現からスタインスプリング表現へ

ここで、写像

$$
\Phi(\rho) = \sum_{k = 0}^{N-1} A_k \rho A_k^{\dagger}
$$

のクラウス表現があり、

$$
\sum_{k = 0}^{N-1} A_k^{\dagger} A_k = \mathbb{I}_{\mathsf{X}}.
$$

を満たすとします。
私たちの目標は $\Phi$ のスタインスプリング表現を見つけることです。

まず行いたいのは、ゴミの系 $\mathsf{G}$ の古典的状態集合が $\{0,\ldots,N-1\}$ となるように $\mathsf{G}$ を選ぶことです。
しかし $(\mathsf{W},\mathsf{X})$ と $(\mathsf{G},\mathsf{Y})$ が同じ大きさであるためには、$n$ が $m N$ を割り切る必要があり、これにより $d = mN/n$ として $\mathsf{W}$ が古典的状態 $\{0,\ldots,d-1\}$ を持つようにできます。

$n$、$m$、$N$ の任意の選び方に対して、$mN/n$ が整数になるとは限らないので、実際には $\mathsf{G}$ の古典的状態集合が $\{0,\ldots,N-1\}$ となるように自由に選べるわけではありません。
しかし、$k$ の追加の値を望むだけ用意し、それらに対して $A_k = 0$ と選ぶことで、クラウス表現において $N$ を常に任意に増やすことができます。

したがって、$mN/n$ が整数であること——これは $N$ が $n/\operatorname{gcd}(n,m)$ の倍数であることと等価です——を暗黙のうちに仮定すれば、$\mathsf{G}$ の古典的状態集合が $\{0,\ldots,N-1\}$ となるように自由に選べます。
特に、$N = nm$ である場合には、$\mathsf{W}$ が $m^2$ 個の古典的状態を持つように選べます。

残るは $U$ を選ぶことであり、これは次のパターンに合わせることで行います。

$$
U =
\begin{pmatrix}
A_{0} & \fbox{?} & \cdots & \fbox{?} \\[1mm]
A_{1} & \fbox{?} & \cdots & \fbox{?} \\[1mm]
\vdots & \vdots & \ddots & \vdots\\[1mm]
A_{N-1} & \fbox{?} & \cdots & \fbox{?}
\end{pmatrix}
$$

明確にしておくと、このパターンはブロック行列を示唆するもので、各ブロック（$A_{0},\ldots,A_{N-1}$ および疑問符で示されたブロックを含む）は $m$ 行 $n$ 列を持ちます。
ブロックの行は $N$ 個あり、これはブロックの列が $d = mN/n$ 個あることを意味します。

より公式的な形で表すと、$U$ を次のように定義します。

$$
\begin{aligned}
U & = \sum_{k=0}^{N-1} \sum_{j=0}^{d-1} \vert k \rangle \langle j \vert \otimes M_{k,j} \\[4mm]
& = \begin{pmatrix}
M_{0,0} & M_{0,1} & \cdots & M_{0,d-1} \\[1mm]
M_{1,0} & M_{1,1} & \cdots & M_{1,d-1} \\[1mm]
\vdots & \vdots & \ddots & \vdots\\[1mm]
M_{N-1,0} & M_{N-1,1} & \cdots & M_{N-1,d-1}
\end{pmatrix}
\end{aligned}
$$

ここで各行列 $M_{k,j}$ は $m$ 行 $n$ 列を持ち、特に $k = 0,\ldots,N-1$ に対して $M_{k,0} = A_k$ とします。

これはユニタリ行列でなければならず、疑問符でラベル付けされたブロック、すなわち $j>0$ に対する $M_{k,j}$ は、このことを念頭に置いて選ばなければなりません——しかし $U$ をユニタリにできるという点を除けば、疑問符でラベル付けされたブロックは証明には何の関係もありません。

$U$ がユニタリであるという懸念をひとまず脇に置き、私たちのスタインスプリング表現について $\mathsf{X}$ の入力状態 $\rho$ が与えられたときの $\mathsf{Y}$ の出力状態を記述する式

$$
\operatorname{Tr}_{\mathsf{G}} \bigl( U (\vert 0\rangle \langle 0 \vert_{\mathsf{W}} \otimes \rho)U^{\dagger}\bigr)
$$

に注目しましょう。
これは次のように書き換えることもできます。

$$
U(\vert 0\rangle\langle 0 \vert \otimes \rho)U^{\dagger}
= U(\vert 0\rangle\otimes\mathbb{I}_{\mathsf{X}}) \rho (\langle 0\vert \otimes \mathbb{I}_{\mathsf{X}}) U^{\dagger},
$$

そして $U$ の選び方から、次がわかります。

$$
U(\vert 0\rangle\otimes\mathbb{I}_{\mathsf{X}}) =
\sum_{k = 0}^{N-1} \vert k\rangle \otimes A_k.
$$

したがって次が得られます。

$$
U(\vert 0\rangle\langle 0 \vert \otimes \rho)U^{\dagger}
= \sum_{j,k = 0}^{N-1} \vert k\rangle\langle j\vert \otimes A_k \rho A_j^{\dagger},
$$

よって

$$
\begin{aligned}
\operatorname{Tr}_{\mathsf{G}} \bigl( U (\vert 0\rangle \langle 0 \vert_{\mathsf{W}} \otimes \rho) U^{\dagger}\bigr)
& = \sum_{j,k = 0}^{N-1} \operatorname{Tr}\bigl(\vert k\rangle\langle j\vert\bigr) \, A_k \rho A_j^{\dagger} \\
& = \sum_{k = 0}^{N-1} A_k \rho A_k^{\dagger} \\
& = \Phi(\rho).
\end{aligned}
$$

したがって、写像 $\Phi$ の正しい表現が得られました。残るは、$U$ をユニタリに選べることを検証することです。

上のパターンに従って選んだときの $U$ の最初の $n$ 列を考えます。
これらの列だけをとると、ブロック行列

$$
\begin{pmatrix}
A_0\\[1mm]
A_1\\[1mm]
\vdots\\[1mm]
A_{N-1}
\end{pmatrix}.
$$

が得られます。
列は $n$ 個あり、$\mathsf{X}$ の各古典的状態に1つ対応します。ベクトルとして、これらの列を各 $a\in\Sigma$ に対して $\vert \gamma_a \rangle$ と名付けましょう。
上のブロック行列表現に対応させられる、これらのベクトルの式は次のとおりです。

$$
\vert \gamma_a\rangle = \sum_{k = 0}^{N-1} \vert k\rangle \otimes A_k \vert a \rangle
$$

さて、これらのベクトルのうち任意の2つ、すなわち任意の $a,b\in\Sigma$ の選び方に対応するものの間の内積を計算しましょう。

$$
\langle \gamma_a \vert \gamma_b \rangle =
\sum_{j,k = 0}^{N-1} \langle k \vert j \rangle \, \langle a \vert A_k^{\dagger} A_j \vert b\rangle =
\langle a \vert \Biggl( \sum_{k = 0}^{N-1} A_k^{\dagger} A_k \Biggr) \vert b\rangle
$$

仮定

$$
\sum_{k = 0}^{m-1} A_k^{\dagger} A_k = \mathbb{I}_{\mathsf{X}}
$$

により、$n$ 個の列ベクトル $\{\vert\gamma_a\rangle\,:\,a\in\Sigma\}$ が正規直交集合をなすと結論できます。

$$
\langle \gamma_a \vert \gamma_b \rangle = \begin{cases}
1 & a = b\\
0 & a\neq b
\end{cases}
$$

がすべての $a,b\in\Sigma$ に対して成り立ちます。

これは、$U$ がユニタリ行列になるように残りの列を埋めることが可能であることを意味します。
特に、残りの列を選ぶのにグラム・シュミットの直交化過程を用いることができます。
同様のことは、「量子情報の基礎」の *量子回路* の講義で、状態識別問題の文脈で行いました。

## スタインスプリング表現から定義へ戻る

最後の含意は 4 $\Rightarrow$ 1 です。
すなわち、系の組 $(\mathsf{W},\mathsf{X})$ を組 $(\mathsf{G},\mathsf{Y})$ に変換するユニタリ操作があると仮定し、私たちの目標は、写像

$$
\Phi(\rho) = \operatorname{Tr}_{\mathsf{G}} \bigl( U (\vert 0\rangle \langle 0 \vert_{\mathsf{W}} \otimes \rho)U^{\dagger}\bigr)
$$

が有効なチャネルであると結論することです。
その形から、$\Phi$ が線形であることは明らかであり、残るはそれが常に密度行列を密度行列に変換することを検証することです。
これはかなり単純で、要点はすでに議論しました。

特に、複合系 $(\mathsf{Z},\mathsf{X})$ の密度行列 $\sigma$ から始め、そこに追加の作業用の系 $\mathsf{W}$ を加えると、確かに密度行列が残ります。
便宜上、系を $(\mathsf{W},\mathsf{Z},\mathsf{X})$ の順に並べ替えると、この状態は次のように書けます。

$$
\vert 0\rangle\langle 0\vert_{\mathsf{W}} \otimes \sigma.
$$

次にユニタリ操作 $U$ を適用します。すでに議論したように、これは有効なチャネルであり、したがって密度行列を密度行列に写します。
最後に、密度行列の部分トレースは別の密度行列です。

これを別の言い方で言えば、まず次のそれぞれが有効なチャネルであることに注目します。

1. 初期化された作業用の系を導入すること。
2. ユニタリ操作を実行すること。
3. ある系をトレースアウトすること。

そして最後に、チャネルの任意の合成は別のチャネルです——これは定義から直ちに従いますが、それ自体として注目に値する事実でもあります。

これで最後の含意の証明が完了し、したがってこの節の冒頭に挙げた4つの命題の同値性が確立されました。
