# はじめに
量子情報の一般的な定式化において、量子状態に対する操作は、*チャネル*と呼ばれる特別な種類の写像によって表現されます。これには、ユニタリーゲートや量子回路に対応する有用な操作だけでなく、ノイズとして見なされ、できれば避けたい操作も含まれます。また、測定もチャネルとして記述することができますが、これについては次のレッスンで扱います。
要するに、（理想化された意味で）物理的に実現可能なあらゆる *離散時間*での状態変化は、チャネルによって記述することができます。（*離散時間*の変化とは、無限に短い時間区間で記述される連続的な過程とは異なり、互いに区別された二つの時刻の間で生じる変化のことを指します。）

ここで*チャネル*という用語は情報理論に由来します。情報理論は、（他の多くのこととともに）ノイズを含む*通信路*がどれだけ情報を運べるかを研究する分野です。この文脈では、量子チャネルは、ある量子状態を送信したときに受信側で得られる量子状態を指定するものと考えることができます。たとえば、何らかの量子ネットワークを通して量子状態を送る場合がその例です。しかし、この用語は歴史的な動機を反映したものであり、現在ではより一般的な意味で使われていることを理解しておくべきです。実際には、量子チャネルによって非常に多様な対象を記述することができます。たとえば複雑な量子計算もチャネルとして表現できますが、それらは通信とは何の関係もなく、通信の文脈で自然に現れることもほとんどありません。

このレッスンではまず、チャネルの基本的な性質と、いくつかの簡単な例について説明します。その後、このレッスンの後半で、チャネルを数学的に表現するための3つの異なる方法を学びます。これらの表現方法は見かけ上は異なりますが、いずれもチャネルを数学的に同等な形で特徴付けていることを確認します。
