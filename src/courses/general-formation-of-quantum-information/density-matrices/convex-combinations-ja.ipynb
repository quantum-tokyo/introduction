# 密度行列の凸結合

## 密度行列の確率的選択

密度行列の重要な特徴の一つは、量子状態の*確率的な選択*が、それらに対応する密度行列の*凸結合*として表されることです。

例えば、系 $\mathsf{X}$ の量子状態を表す2つの密度行列 $\rho$ と $\sigma$ があり、確率 $p$ で状態 $\rho$、確率  $1 - p$ で状態 $\sigma$ を準備するとします。このとき得られる量子状態は、次の密度行列で表されます：


$$
p \rho + (1 - p) \sigma.
$$

より一般に、密度行列 $\rho_0,\ldots,\rho_{m-1}$ で表される $m$ 個の量子状態があり、確率ベクトル $(p_0,\ldots,p_{m-1})$ に従って、状態 $\rho_k$ を確率 $p_k$ で準備するとします。このとき得られる状態は、次の密度行列で表されます：

$$
\sum_{k = 0}^{m-1} p_k \rho_k.
$$

これは、密度行列 $\rho_0,\ldots,\rho_{m-1}$ の*凸結合*です。


同様に、量子状態ベクトル $\vert\psi_0\rangle,\ldots,\vert\psi_{m-1}\rangle$ があり、各 $k\in\{0,\ldots,m-1\}$ に対して確率 $p_k$ で状態 $\vert\psi_k\rangle$ に系を準備する場合、得られる状態は次の密度行列で表されます：


$$
\sum_{k = 0}^{m-1} p_k \vert\psi_k\rangle\langle\psi_k\vert.
$$


例えば、1量子ビットが確率 $1/2$ で $\vert 0\rangle$、確率 $1/2$  で $\vert + \rangle$  に準備されるとします。このときの密度行列は

$$
\frac{1}{2} \vert 0\rangle\langle 0 \vert + \frac{1}{2} \vert +\rangle\langle + \vert
= \frac{1}{2} \begin{pmatrix} 1 & 0\\[1mm] 0 & 0 \end{pmatrix} + \frac{1}{2} \begin{pmatrix} \frac{1}{2} & \frac{1}{2}\\[2mm] \frac{1}{2} & \frac{1}{2} \end{pmatrix}
= \begin{pmatrix} \frac{3}{4} & \frac{1}{4}\\[2mm]
\frac{1}{4} & \frac{1}{4}
\end{pmatrix}.
$$

一方で、量子情報の簡略化された定式化においては、このように量子状態ベクトルを単純に平均する方法はうまくいきません。

例えば、次のベクトル

$$
\frac{1}{2} \vert 0\rangle + \frac{1}{2} \vert + \rangle
= \frac{1}{2} \begin{pmatrix}1\\[1mm] 0\end{pmatrix} + \frac{1}{2} \begin{pmatrix}\frac{1}{\sqrt{2}}\\[2mm]\frac{1}{\sqrt{2}}\end{pmatrix}
= \begin{pmatrix}\frac{2 + \sqrt{2}}{4}\\[2mm]\frac{\sqrt{2}}{4}\end{pmatrix}
$$

は、そのユークリッドノルムが $1$ ではないため、正しい量子状態ベクトルではありません。さらに極端な例として、任意の量子状態ベクトル $\vert\psi\rangle$ を固定し、確率 $1/2$ で $\vert\psi\rangle$、確率 $1/2$ で $-\vert\psi\rangle$ を取るとします。これらはグローバル位相が異なるだけなので、実際には同じ状態を表しています。しかし、それらを平均するとゼロベクトルになってしまい、これは有効な量子状態ベクトルではありません。

## 完全混合状態

量子ビットの状態を $\vert 0\rangle$ または $\vert 1\rangle$  にランダムに設定し、それぞれ確率 $1/2$で選ぶとします。このとき得られる状態を表す密度行列は次の通りです：

$$
\frac{1}{2} \vert 0\rangle\langle 0\vert + \frac{1}{2} \vert 1\rangle\langle 1\vert
= \frac{1}{2}
\begin{pmatrix}
1 & 0\\[1mm]
0 & 0
\end{pmatrix} + \frac{1}{2}
\begin{pmatrix}
0 & 0\\[1mm]
0 & 1
\end{pmatrix}
= \begin{pmatrix}
\frac{1}{2} & 0\\[1mm]
0 & \frac{1}{2}
\end{pmatrix}
= \frac{1}{2} \mathbb{I}
$$

（ここで $\mathbb{I}$ は $2\times 2$の単位行列を表します。）
これは*完全混合状態*と呼ばれる特別な状態です。これは量子ビットの状態としては完全に不確定であることを表しており、確率論における一様にランダムなビットに相当します。


次に手続きを変更してみましょう：$\vert 0\rangle$ と $\vert 1\rangle$ の代わりに、$\vert + \rangle$ と $\vert - \rangle$ を使います。このときの密度行列も同様に計算できます：

$$
\frac{1}{2} \vert +\rangle\langle +\vert + \frac{1}{2} \vert -\rangle\langle -\vert
= \frac{1}{2}
\begin{pmatrix}
\frac{1}{2} & \frac{1}{2}\\[2mm]
\frac{1}{2} & \frac{1}{2}
\end{pmatrix}
+ \frac{1}{2}
\begin{pmatrix}
\frac{1}{2} & -\frac{1}{2}\\[2mm]
-\frac{1}{2} & \frac{1}{2}
\end{pmatrix}
=
\begin{pmatrix}
\frac{1}{2} & 0\\[2mm]
0 & \frac{1}{2}
\end{pmatrix}
= \frac{1}{2} \mathbb{I}
$$

状態を変えたにもかかわらず、先ほどと同じ密度行列になります。
実際、$\vert 0\rangle$ と $\vert 1\rangle$ の代わりに*任意の*直交する2つの量子ビット状態ベクトルを用いても、同じ結果 — 完全混合状態 — が得られます。

これはバグではなく、重要な特徴です！
実際、どちらの方法でも全く同じ状態が得られます。
つまり、生成された量子ビットを測定しても（統計的に見ても）この2つの手続きを区別することはできません。
これらは単に同じ状態を準備する異なる方法にすぎません。

このことが妥当であることは、次のように考えると確認できます。
2つの状態集合 $\{\vert 0\rangle,\vert 1\rangle\}$ または $\{\vert +\rangle,\vert -\rangle\}$ のどちらかからランダムに状態が選ばれるとき、そこから何が分かるかを考えます。
簡単のため、量子ビットにユニタリー演算 $U$ を施し、その後、標準基底で測定するとします。

最初の手続きでは、量子ビットの状態は、$\{\vert 0\rangle,\vert 1\rangle\}$からランダムに選ばれるとします。
状態が $\vert 0\rangle$ のとき、測定結果 $0$ と $1$ の確率はそれぞれ

$$
\vert \langle 0 \vert U \vert 0 \rangle \vert^2
\quad\text{and}\quad
\vert \langle 1 \vert U \vert 0 \rangle \vert^2
$$

状態が $\vert 1\rangle$ のとき、測定結果 $0$ と $1$ の確率はそれぞれ

$$
\vert \langle 0 \vert U \vert 1 \rangle \vert^2
\quad\text{and}\quad
\vert \langle 1 \vert U \vert 1 \rangle \vert^2.
$$
となります。

それぞれが確率 $1/2$ で起こるため、結果 $0$ の確率は

$$
\frac{1}{2}\vert \langle 0 \vert U \vert 0 \rangle \vert^2
+ \frac{1}{2}\vert \langle 0 \vert U \vert 1 \rangle \vert^2
$$

結果$1$ の確率は

$$
\frac{1}{2}\vert \langle 1 \vert U \vert 0 \rangle \vert^2
+ \frac{1}{2}\vert \langle 1 \vert U \vert 1 \rangle \vert^2.
$$

となり、これらはどちらも $1/2$ となります。
これは線形代数の次の事実（ピタゴラスの定理の一般化）から分かります：



$\{\vert\psi_1\rangle,\ldots,\vert\psi_n\rangle\}$ が（実数または複素数の）ベクトル空間 $\mathcal{V}$ の正規直交基底であるとする。このとき、任意のベクトル $\vert \phi\rangle \in \mathcal{V}$ に対して、
$\vert \langle \psi_1\vert\phi\rangle\vert^2 + \cdots + \vert \langle \psi_n \vert \phi \rangle\vert^2 = \| \vert\phi\rangle \|^2$ が成り立つ。


この定理を用いることで、確率を次のように求めることができます。
結果 $0$ を得る確率は 

$$
\begin{aligned}
\frac{1}{2}\vert \langle 0 \vert U \vert 0 \rangle \vert^2
+ \frac{1}{2}\vert \langle 0 \vert U \vert 1 \rangle \vert^2
& = \frac{1}{2} \Bigl(
\vert \langle 0 \vert U \vert 0 \rangle \vert^2
+ \vert \langle 0 \vert U \vert 1 \rangle \vert^2
\Bigr) \\[2mm]
& = \frac{1}{2} \Bigl(
\vert \langle 0 \vert U^{\dagger} \vert 0 \rangle \vert^2
+ \vert \langle 1 \vert U^{\dagger} \vert 0 \rangle \vert^2
\Bigr)\\[2mm]
& = \frac{1}{2} \bigl\| U^{\dagger} \vert 0 \rangle \bigr\|^2
\end{aligned}
$$
となります。
同様に、結果 $1$を得る確率は

$$
\begin{aligned}
\frac{1}{2}\vert \langle 1 \vert U \vert 0 \rangle \vert^2
+ \frac{1}{2}\vert \langle 1 \vert U \vert 1 \rangle \vert^2
& = \frac{1}{2} \Bigl(
\vert \langle 1 \vert U \vert 0 \rangle \vert^2
+ \vert \langle 1 \vert U \vert 1 \rangle \vert^2
\Bigr) \\[2mm]
& = \frac{1}{2} \Bigl(
\vert \langle 0 \vert U^{\dagger} \vert 1 \rangle \vert^2
+ \vert \langle 1 \vert U^{\dagger} \vert 1 \rangle \vert^2
\Bigr)\\[2mm]
& = \frac{1}{2} \bigl\| U^{\dagger} \vert 1 \rangle \bigr\|^2.
\end{aligned}
$$
となります。

ここで、 $U$ はユニタリーであるため、 $U^{\dagger}$ もまたユニタリーです。したがって、 $U^{\dagger} \vert 0 \rangle$と$U^{\dagger} \vert 1 \rangle$ はいずれも長さ1のベクトル（単位ベクトル）になります。
よって、これらの確率はいずれも $1/2$になります。
つまり、どのように $U$ を選んでも、測定によって得られるのは一様ランダムなビットに過ぎません。

同様の検証は、 $\vert 0\rangle$ と$\vert 1\rangle$の代わりに任意の正規直交な状態の組についても行えます。
例えば、 $\{\vert + \rangle, \vert - \rangle\}$ が正規直交基底であることから、2つ目の手続きにおいて測定結果$0$ を得る確率は

$$
\frac{1}{2}\vert \langle 0 \vert U \vert + \rangle \vert^2
+ \frac{1}{2}\vert \langle 0 \vert U \vert - \rangle \vert^2
= \frac{1}{2} \bigl\| U^{\dagger} \vert 0 \rangle \bigr\|^2 = \frac{1}{2}
$$

となり、結果 $1$ を得る確率は 

$$
\frac{1}{2}\vert \langle 1 \vert U \vert + \rangle \vert^2
+ \frac{1}{2}\vert \langle 1 \vert U \vert - \rangle \vert^2
= \frac{1}{2} \bigl\| U^{\dagger} \vert 1 \rangle \bigr\|^2 = \frac{1}{2}.
$$
となります。
特に、$\vert 0\rangle$ と$\vert 1\rangle$ の場合とまったく同じ測定統計が得られることが分かります。

## 確率的状態

古典状態は密度行列によって表すことができます。
特に、系 $\mathsf{X}$ の各古典状態 $a$ に対して、密度行列

$$ \rho = \vert a\rangle \langle a \vert $$

は、$\mathsf{X}$ が確定的に古典状態 $a$ にあることを表します。
量子ビットの場合には、

$$
\vert 0\rangle \langle 0 \vert = \begin{pmatrix}1 & 0 \\ 0 & 0\end{pmatrix}
\quad\text{and}\quad
\vert 1\rangle \langle 1 \vert = \begin{pmatrix}0 & 0 \\ 0 & 1\end{pmatrix}
$$
となります。

一般には、考えている古典状態に対応する位置の対角成分に $1$ が1つ入り、それ以外の成分はすべて 0 になります。

これらの密度行列の凸結合を取ることで、確率的状態を表すことができます。
簡単のため、古典状態の集合が$\{0,\ldots,n-1\}$であるとします。各$a\in\{0,\ldots,n-1\}$ に対して、$\mathsf{X}$ が状態 $a$ に確率 $p_a$ であるとすると、対応する密度行列は

$$
\rho = \sum_{a = 0}^{n-1} p_a \vert a\rangle \langle a \vert
= \begin{pmatrix}
p_0 & 0 & \cdots & 0\\
0 & p_1 & \ddots & \vdots\\
\vdots & \ddots & \ddots & 0\\
0 & \cdots & 0 & p_{n-1}
\end{pmatrix}.
$$
となります。
逆に言えば、任意の対角な密度行列は、その対角成分を読み取ることで対応する確率分布（確率ベクトル）として自然に解釈できます。

ただし明確にしておくと、密度行列が対角であるからといって、必ずしもその系が古典的であるとか、古典状態のランダムな選択によって準備されたとは限りません。
そうではなく、その状態は「古典状態のランダム選択によって得られた*可能性がある*」ということを意味します。

確率的状態が対角な密度行列で表されるという事実は、この講義の冒頭で述べた直感とも一致しています。すなわち、非対角成分は、その行と列に対応する2つの古典状態がどの程度量子重ね合わせにあるかを表しています。
ここではすべての非対角成分が 0 であるため、単なる古典的なランダム性だけが存在し、量子重ね合わせは全く含まれていません。

## 密度行列とスペクトル定理

純粋状態の凸結合

$$
\rho = \sum_{k = 0}^{m-1} p_k \vert \psi_k\rangle \langle \psi_k \vert
$$

を取ると、密度行列が得られることを見てきました。

実際には、任意の密度行列 $\rho$  は、このような純粋状態の凸結合として表すことができます。
すなわち、ある単位ベクトルの集合$\{\vert\psi_0\rangle,\ldots,\vert\psi_{m-1}\rangle\}$ と確率ベクトル$(p_0,\ldots,p_{m-1})$ が存在して、上の式が成り立ちます。

さらに、数 $m$ は考えている系の古典状態の数と一致するように常に選ぶことができ、また量子状態ベクトルは互いに直交するように選ぶこともできます。
これは、"量子アルゴリズムの基礎" コースで扱ったスペクトル定理から導かれます。
以下に便宜のため、そのスペクトル定理を改めて述べます。


スペクトル定理: $M$ を $n\times n$ の *正規* 複素行列とする。
このとき、$n$次元複素ベクトル空間の正規直交基底$\{\vert\psi_0\rangle,\ldots,\vert\psi_{n-1}\rangle \}$ と複素数 $\lambda_0,\ldots,\lambda_{n-1}$ が存在して、
$$
M = \lambda_0 \vert \psi_0\rangle\langle \psi_0\vert + \cdots + \lambda_{n-1} \vert \psi_{n-1}\rangle\langle \psi_{n-1}\vert.
$$と表すことができる。


(行列 $M$ が *正規* であるとは、$M^{\dagger} M = M M^{\dagger}.$ を満たすことを思い出してください。言い換えると、正規行列とは自分自身の共役転置と可換な行列のことです。)

密度行列は常にエルミートであり、したがって正規行列でもあるため、任意の密度行列 $\rho$  に対してスペクトル定理を適用することができます。
これにより、次のように書くことができます。

$$
\rho = \lambda_0 \vert \psi_0\rangle\langle \psi_0\vert + \cdots + \lambda_{n-1} \vert \psi_{n-1}\rangle\langle \psi_{n-1}\vert
$$

ある正規直交基底 $\{\vert\psi_0\rangle,\ldots,\vert\psi_{n-1}\rangle\}$ に対して。
あとは、 $(\lambda_0,\ldots,\lambda_{n-1})$ が確率ベクトルであることを確認すればよく、その後で必要に応じて $(p_0,\ldots,p_{n-1})$ と書き直すことができます。

これらの数 $\lambda_0,\ldots,\lambda_{n-1}$ は $\rho$ の固有値であり、$\rho$ は半正定値であるため、これらは非負の実数でなければなりません。
さらに、  $\rho$ のトレースが$1$であることから、$\lambda_0 + \cdots + \lambda_{n-1} = 1$ が成り立つことが分かります。
詳細を確認することで、トレースに関する次の重要で非常に有用な性質を指摘することができます。


定理（トレースの巡回性）: 行列の積$AB$ が正方行列になるような任意の2つの行列$A$、 $B$ に対して、 $\operatorname{Tr}(AB) = \operatorname{Tr}(BA)$ が成り立つ。


この定理は、$A$や $B$ 自体が正方行列でない場合でも成立することに注意してください。
すなわち、ある正の整数 $n$ ,$m$ に対して、$A$ が $n\times m$行列、$B$ が $m\times n$ 行列である場合でも、$AB$は$n\times n$ の正方行列、$BA$ は$m\times m$の正方行列となり、この関係は成り立ちます。

特に、$A$を列ベクトル $\vert\phi\rangle$、 $B$を行ベクトル$\langle \phi\vert$ とすると、

$$
\operatorname{Tr}\bigl(\vert\phi\rangle\langle\phi\vert\bigr)
= \operatorname{Tr}\bigl(\langle\phi\vert\phi\rangle\bigr) = \langle\phi\vert\phi\rangle.
$$
が成り立つことが分かります。

2つ目の等式は、 $\langle\phi\vert\phi\rangle$がスカラーであり、$1\times 1$行列（唯一の成分がその値）とみなせるため、そのトレースがその値に等しいことから従います。
この事実を用いると、トレースの線形性により、 $\lambda_0 + \cdots + \lambda_{n-1} = 1$ が成り立つことが分かります。

$$
\begin{gathered}
1 = \operatorname{Tr}(\rho) =
\operatorname{Tr}\bigl(\lambda_0 \vert \psi_0\rangle\langle \psi_0\vert + \cdots + \lambda_{n-1} \vert \psi_{n-1}\rangle\langle \psi_{n-1}\vert\bigr)\\[2mm]
= \lambda_0 \operatorname{Tr}\bigl(\vert \psi_0\rangle\langle \psi_0\vert\bigr) + \cdots + \lambda_{n-1} \operatorname{Tr}\bigl(\vert \psi_{n-1}\rangle\langle \psi_{n-1}\vert\bigr)
= \lambda_0 + \cdots + \lambda_{n-1}
\end{gathered}
$$

別の方法として、正方行列のトレースは（正規行列であるかどうかに関わらず）その固有値の和に等しいという事実からも、同じ結論に到達できます。

以上より、任意の密度行列 $\rho$は純粋状態の凸結合として表せることが分かりました。さらに、その純粋状態を互いに*直交*するように選ぶことも可能です。
これは特に、必要となる数 $n$ が、系 $\mathsf{X}$の古典状態の数を超える必要がないことを意味します。

一般に、密度行列を純粋状態の凸結合として表す方法は、スペクトル定理によるものに限らず、複数存在することに注意が必要です。
このことは、先ほどの例からも分かります。

$$
\frac{1}{2} \vert 0\rangle\langle 0 \vert + \frac{1}{2} \vert +\rangle\langle + \vert
= \begin{pmatrix}
\frac{3}{4} & \frac{1}{4}\\[2mm]
\frac{1}{4} & \frac{1}{4}
\end{pmatrix}
$$

これはこの行列のスペクトル分解ではありません。なぜなら、 $\vert 0\rangle$ と$\vert + \rangle$は直交していないからです。
次がスペクトル分解の一例です：

$$
\begin{pmatrix}
\frac{3}{4} & \frac{1}{4}\\[2mm]
\frac{1}{4} & \frac{1}{4}
\end{pmatrix}
= \cos^2(\pi/8) \vert \psi_{\pi/8} \rangle \langle \psi_{\pi/8}\vert +
\sin^2(\pi/8) \vert \psi_{5\pi/8} \rangle \langle \psi_{5\pi/8}\vert
$$

ここで$\vert \psi_{\theta} \rangle = \cos(\theta)\vert 0\rangle + \sin(\theta)\vert 1\rangle$です。
固有値は、見覚えのある形をしています：

$$
\cos^2(\pi/8) = \frac{2+\sqrt{2}}{4} \approx 0.85 \quad\text{and}\quad \sin^2(\pi/8) = \frac{2-\sqrt{2}}{4} \approx 0.15.
$$

固有ベクトルは明示的に次のように書けます

$$
\begin{aligned}
  \vert\psi_{\pi/8}\rangle & =
  \frac{\sqrt{2 + \sqrt{2}}}{2}\vert 0\rangle + \frac{\sqrt{2 - \sqrt{2}}}{2}\vert 1\rangle \\[3mm]
  \vert\psi_{5\pi/8}\rangle & =
  -\frac{\sqrt{2 - \sqrt{2}}}{2}\vert 0\rangle + \frac{\sqrt{2 + \sqrt{2}}}{2}\vert 1\rangle
\end{aligned}
$$

別の、より一般的な例を考えます。 $\vert \phi_0\rangle,\ldots,\vert \phi_{99} \rangle$ を1量子ビットの状態を表す量子状態ベクトルとし、これらは任意に選ばれている（つまり特別な関係は仮定しない）とします。
このとき、これら$100$個の状態の中から一様ランダムに1つを選ぶことで得られる状態を考えると、

$$\rho = \frac{1}{100} \sum_{k = 0}^{99} \vert \phi_k\rangle\langle \phi_k \vert$$

となります。
量子ビットについて考えているので、密度行列 $\rho$ は $2\times 2$ 行列です。したがってスペクトル定理より、これを

$$
\rho = p \vert\psi_0\rangle\langle\psi_0\vert + (1 - p) \vert\psi_1\rangle\langle\psi_1\vert
$$

という形でも書くことができます。ここで$p\in[0,1]$ は実数であり、$\{\vert\psi_0\rangle,\vert\psi_1\rangle\}$ は正規直交基底です。— しかし、このような表現が存在するからといって、$\rho$ を100個の純粋状態の平均として表すことができないわけではありません。もちろん、そのように表すことも可能です。
